# Восстановление золота из руды

Необходимо подготовить прототип модели машинного обучения для «Цифры». Компания разрабатывает решения для эффективной работы промышленных предприятий. Модель поможет оптимизировать производство, чтобы не запускать предприятие с убыточными характеристиками.

Модель должна предсказать коэффициент восстановления золота из золотосодержащей руды. Будут использованы данные с параметрами добычи и очистки. Данные находятся в трех файлах - исходные данные, обучающая и тестовая выборка. Некоторые параметры недоступны, потому что замеряются и/или рассчитываются значительно позже. Из-за этого в тестовой выборке отсутствуют некоторые признаки, которые могут быть в обучающей. Также в тестовом наборе нет целевых признаков. **Описание данных**

***Технологический процесс***

- Rougher feed — исходное сырье  
- Rougher additions (или reagent additions) — флотационные реагенты:  
Xanthate, Sulphate, Depressant  
  - Xanthate — ксантогенат (промотер, или активатор флотации);  
  - Sulphate — сульфат (на данном производстве сульфид натрия);  
  - Depressant — депрессант (силикат натрия).  
- Rougher process (англ. «грубый процесс») — флотация
- Rougher tails — отвальные хвосты
- Float banks — флотационная установка
- Cleaner process — очистка
- Rougher Au — черновой концентрат золота
- Final Au — финальный концентрат золота

***Параметры этапов***

- air amount — объём воздуха
- fluid levels — уровень жидкости
- feed size — размер гранул сырья
- feed rate — скорость подачи

***Наименование признаков***

Наименование признаков должно быть такое:

[этап].[тип_параметра].[название_параметра]

Пример: rougher.input.feed_ag

Возможные значения для блока [этап]:
- rougher — флотация
- primary_cleaner — первичная очистка
- secondary_cleaner — вторичная очистка
- final — финальные характеристики

Возможные значения для блока [тип_параметра]:
- input — параметры сырья
- output — параметры продукта
- state — параметры, характеризующие текущее состояние этапа
- calculation — расчётные характеристики

### Проверка расчета эффективности обогащения

Проверим, что эффективность обогащения рассчитана правильно. Вычислим её на обучающей выборке для признака rougher.output.recovery. Найдем MAE между нашими расчётами и значением признака. Эффективность обогащения рассчитывается по формуле:

где:  
C — доля золота в концентрате после флотации/очистки;  
F — доля золота в сырье/концентрате до флотации/очистки;  
T — доля золота в отвальных хвостах после флотации/очистки.  

```python
def recovery (C, F, T):
    return C * (F - T) / (F * (C - T)) * 100

print("MAE между расчётной эффективностью обогащения и значением признака из трэйн датасета:", 
      round(mean_absolute_error(data_train["rougher.output.recovery"], 
                                recovery (data_train["rougher.output.concentrate_au"], 
                                          data_train["rougher.input.feed_au"], 
                                          data_train["rougher.output.tail_au"]))))
```

**Вывод:** Эффективность обогащения рассчитана верно.

## Модель

### Функция для вычисления итоговой sMAPE

Нужно спрогнозировать сразу две величины:  
эффективность обогащения чернового концентрата rougher.output.recovery;  
эффективность обогащения финального концентрата final.output.recovery.  

Для решения задачи введём новую метрику качества — sMAPE (англ. Symmetric Mean Absolute Percentage Error, «симметричное среднее абсолютное процентное отклонение»).
Метрика sMAPE вычисляется так:

Обозначения:  
yi - Значение целевого признака для объекта с порядковым номером i в выборке, на которой измеряется качество  
ŷi - Значение предсказания для объекта с порядковым номером i, например, в тестовой выборке  
N - Количество объектов в выборке  
Σ - Суммирование по всем объектам выборки (i меняется от 1 до N)

```python
# sMAPE для чернового или финального концентрата
def sMAPE (tar, pred):
    return 1 / len(tar) * sum(abs(tar - pred) / ((abs(tar) + abs(pred)) / 2)) * 100
```

```python
# итоговое sMAPE
def final_sMAPE (sMAPE_rou, sMAPE_fin):
    return 0.25 * sMAPE_rou + 0.75 * sMAPE_fin
```

```python
# для кросс-валидации
sMAPE_scorer = make_scorer(sMAPE, greater_is_better=False)
```

### Обучение моделей

Обучим разные модели и оценим их качество кросс-валидацией. Выберем лучшую модель и проверим её на тестовой выборке. Данные уже разбиты на тренировочную и тестовую выборки, у нас два целевых признака, значит на каждый нужно по модели. Для деревянных моделей можно не делать стандартизацию, нужно удалить столбец с датой из наборов данных, этот столбец не понадобится моделям.

```python
# извлечем признаки для предсказания эффективности обогащения чернового и финального концентрата из трэйн выборки
features_train = data_train.drop(["date", "rougher.output.recovery", "final.output.recovery"], axis=1)

# извлечем целевой признак 1
target_train_rougher = data_train["rougher.output.recovery"]
# извлечем целевой признак 2
target_train_final = data_train["final.output.recovery"]
```

```python
# проверим корректность разбиения данных
for df_tr in [features_train, target_train_rougher, target_train_final]:
    print(df_tr.shape)
```

#### Решающее дерево

```python
%%time

parametrs_dtr = { "max_depth": range (1,5),
                "min_samples_leaf": range(1,5) }

dtr_rou = DecisionTreeRegressor(random_state=42)
grid_model_dtr_rou = GridSearchCV(dtr_rou, parametrs_dtr, scoring = sMAPE_scorer)
grid_model_dtr_rou.fit(features_train, target_train_rougher)
sMAPE_dtr_rou = round(abs(grid_model_dtr_rou.best_score_), 2)
print("sMAPE для чернового концентрата", sMAPE_dtr_rou)

dtr_fin = DecisionTreeRegressor(random_state=42)
grid_model_dtr_fin = GridSearchCV(dtr_fin, parametrs_dtr, scoring = sMAPE_scorer)
grid_model_dtr_fin.fit(features_train, target_train_final)
sMAPE_dtr_fin = round(abs(grid_model_dtr_fin.best_score_), 2)
print("sMAPE для финального концентрата", sMAPE_dtr_fin)

final_sMAPE_dtr = round(final_sMAPE (sMAPE_dtr_rou, sMAPE_dtr_fin), 2)
print("Итоговое sMAPE", final_sMAPE_dtr)
```

#### Случайный лес

```python
%%time

parametrs_rfr = { "n_estimators": range (1,5), 
                "max_depth": range (1,5) }

rfr_rou = RandomForestRegressor(random_state=42)
grid_model_rfr_rou = GridSearchCV(rfr_rou, parametrs_rfr, scoring = sMAPE_scorer)
grid_model_rfr_rou.fit(features_train, target_train_rougher)
sMAPE_rfr_rou = round(abs(grid_model_rfr_rou.best_score_), 2)
print("sMAPE для чернового концентрата", sMAPE_rfr_rou)

rfr_fin = RandomForestRegressor(random_state=42)
grid_model_rfr_fin = GridSearchCV(rfr_fin, parametrs_rfr, scoring = sMAPE_scorer)
grid_model_rfr_fin.fit(features_train, target_train_final)
sMAPE_rfr_fin = round(abs(grid_model_rfr_fin.best_score_), 2)
print("sMAPE для финального концентрата", sMAPE_rfr_fin)

final_sMAPE_rfr = round(final_sMAPE (sMAPE_rfr_rou, sMAPE_rfr_fin), 2)
print("Итоговое sMAPE", final_sMAPE_rfr)
```

#### Линейная регрессия

```python
pipe_rou = make_pipeline(StandardScaler(), LinearRegression())
sMAPE_lr_rou = round(abs(cross_val_score(
    pipe_rou, features_train, target_train_rougher, scoring = sMAPE_scorer, cv=5).mean()), 2)
print("sMAPE для чернового концентрата", sMAPE_lr_rou)

pipe_fin = make_pipeline(StandardScaler(), LinearRegression())
sMAPE_lr_fin = round(abs(cross_val_score(
    pipe_fin, features_train, target_train_final, scoring = sMAPE_scorer, cv=5).mean()), 2)
print("sMAPE для чернового концентрата", sMAPE_lr_fin)

final_sMAPE_lr = round(final_sMAPE (sMAPE_lr_rou, sMAPE_lr_fin), 2)
print("Итоговое sMAPE", final_sMAPE_lr)
```
